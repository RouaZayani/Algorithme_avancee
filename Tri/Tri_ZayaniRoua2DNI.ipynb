<h1 style='color:#26B260'> Tri de selection </h1>

```python
import random
import matplotlib.pyplot as plt
import numpy as np
from time import time
```

```python
temps=[]
t=[]
t_bulle=[]
t_inse=[]
t_rap=[]
t_fus=[]
def table(j):
    for i in range(j):
        x=random.randint(0,j)
        t.append(x)
    return t
```

```python
def tri_selection (v,ordre):
    if ordre=='croissant' :
         for i in range(len(v)):
            ind_min=i
            for j in range(i+1,len(v)):
                if v[j]<v[ind_min]:
                    ind_min=j
                    
            aux=v[i]
            v[i]=v[ind_min]
            v[ind_min]=aux    
    elif(ordre=='decroissant'):
          for i in range(len(v)):
            ind_max=i
            for j in range(i+1,len(v)):
                if v[j]>v[ind_max]:
                    ind_max=j
                    
            aux=v[i]
            v[i]=v[ind_max]
            v[ind_max]=aux
    fin=time()
    d=fin-debut
    temps.append(d)
```

```python
v=table(40000)
debut=time()
tri_selection(v,'croissant')
```

```python
n=[20000,25000,30000,35000]
plt.figure(figsize=(12,10))
plt.plot(n,temps,"red")
plt.title("Tri Selection")
plt.xlabel("N ")
plt.ylabel("Temps CPU en seconde ")
plt.grid()
plt.show()
```

<h1 style='color:#26B260'> Tri à bulle </h1>

```python
def tri_bulle(v,ordre):
    permut=True
    p=len(v)-1
    if ordre=='croissant':
        while permut :
            permut=False
            for i in range(p):
                if v[i]>v[i+1]:
                    x=v[i]
                    v[i]=v[i+1]
                    v[i+1]=x
                    permut=True
            p-=1
    elif ordre=='decroissant':
        while permut :
            permut=False
            for i in range(p):
                if v[i]<v[i+1]:
                    x=v[i]
                    v[i]=v[i+1]
                    v[i+1]=x
                    permut=True
            p-=1
    fin=time()
    d=fin-deb
    t_bulle.append(d)
```

```python
for i in range(len(n)):
    a=table(n[i])
    deb=time()
    tri_bulle(a,'decroissant')
```

<h1 style='color:#26B260'> Tri par fusion </h1>

```python
def tri_fusion(v) :
    dd=time()
    if len(v) <= 1:
        return v
    milieu = len(v) // 2
    v1 = v[:milieu]
    v2 = v[milieu:]
    v1=tri_fusion(v1)
    v2=tri_fusion(v2)
    i=0
    j=0
    k=0
    while (i<len(v1)) and (j<len(v2)):
        if v1[i] <= v2[j]:
            v.append(v1[i])
            i+=1
        else:
            v.append(v2[j])
            j=j+1
    if (i<len(v1)):
        v.extend(v1[i:])
    if (j<len(v2)):
        v.extend(v2[j:])
    fin=time()
    e=fin-dd
    t_fus.append(e)
```

<h1 style='color:#26B260'> Tri rapide </h1> 

```python
def choisi_pivot(v,pivot):
    v1=[]
    v2=[]
    v3=[]
    for i in range(len(v)):
        x=v[i]
        if (x<pivot):
            v1.append(x)
        elif(x>pivot):
            v3.append(x)
        else:
            v2.append(x)
    return (v1,v2,v3)  
def tri_rapide(v):
    if len(v)<=1:
        return v
    else:
        pivot=v[0]
        (v1,v2,v3)=choisi_pivot(v,pivot)
        v1=tri_rapide(v1)
        v3=tri_rapide(v3)
    return list(v1+v2+v3)
```

```python
n2=[20000,25000,30000,35000,40000,45000,50000,55000,60000,65000]
for i in range(len(n2)):
    v=table(n2[i])
    d=time()
    tri_rapide(v)
    t_rap.append(time()-d)
```
